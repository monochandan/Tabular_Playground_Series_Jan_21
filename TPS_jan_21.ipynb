```python
### replace outliers by median value
def replace_outliers(data):
    for col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q1 - Q3
        median_ = data[col].median()
        data.loc[((data[col]< Q1 - 1.5*IQR ) | (data[col] > Q1 - 1.5*IQR )), col] = median_ ## if value is outside of the interquertile range replace by the median value
        
    return data
```

```python
df_train = replace_outliers(df_train)
df_test = replace_outliers(df_test)
```

```python
x = df_train.drop(['target'], axis = 1)
y = df_train.target
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.18, random_state = 33)
```

GRIDSEARCHCV  used for select best value for hyperparameters

```python
import lightgbm as lgb
LGB = lgb.LGBMRegressor(random_state = 33, n_estimators=4800, min_data_oer_group = 5, boosting_type = 'gbdt', num_leaves=246,
                        max_dept = -1, learning_rate = 0.005, subsample_for_bin = 200000,
                        lambda_l1 = 1.074622455507616e-05, lambda_l2 = 2.0521330798729704e-06,
                        n_jobs = -1, cat_smooth = 1.0, importance_type = 'split', metric = 'rmse', min_child_samples = 20, min_gain_to_split = 0.0,
                        feature_fraction = 0.5, bagging_freq = 6, min_sum_hessian_in_leaf = 0.001,  min_data_in_leaf = 100, bagging_fraction= 0.82063411)
LGB.fit(x_train,y_train)
```

```python
pred_LGB = LGB.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mse_LGB = np.sqrt(mean_squared_error(y_test, pred_LGB))
mse_LGB
```
